What is the average age of men?

```python
round(df.loc[df['sex'] == 'Male', 'age'].mean(), 1)
```

What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

```python
df['education'].unique()
```

```python
advanced_education = df.loc[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
```

```python
perc = (len(advanced_education[advanced_education['salary'] == '>50K']) / len(advanced_education))*100
round(perc, 1)
```

What percentage of people without advanced education make more than 50K?

```python
school_education = df.loc[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
```

```python
perc = (len(school_education[school_education['salary'] == '>50K']) / len(school_education))*100
round(perc, 1)
```

What is the minimum number of hours a person works per week?

```python
df['hours-per-week'].min()
```

What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
min_h_workers = df.loc[df["hours-per-week"] == df['hours-per-week'].min()]
```

```python
perc = (len(min_h_workers[min_h_workers['salary'] == '>50K']) / len(min_h_workers))*100
round(perc, 1)
```

What country has the highest percentage of people that earn >50K and what is that percentage?

```python
high_perc = (df[df['salary'] == '>50K']['native-country'].value_counts() / df['native-country'].value_counts())*100
high_perc.idxmax()
```

Identify the most popular occupation for those who earn >50K in India.

```python
Indians = df[df['native-country'] == 'India']
Indians = Indians[Indians['salary'] == '>50K']
popular_occupation = Indians['occupation'].value_counts().idxmax()
popular_occupation
```
